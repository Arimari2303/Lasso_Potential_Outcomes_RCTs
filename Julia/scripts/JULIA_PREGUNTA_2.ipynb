```julia
# ========================================
# LASSO para Análisis de Alfabetización Femenina en India
# Traducción de Python a Julia
# ========================================

# Instalar paquetes necesarios (ejecutar una sola vez)
# using Pkg
# Pkg.add(["DataFrames", "XLSX", "Plots", "GLMNet", "StatsBase", "Statistics", 
#          "MLJ", "MLJLinearModels", "Random", "LinearAlgebra"])

# Cargar librerías
using DataFrames
using XLSX
using Plots
using GLMNet
using StatsBase
using Statistics
using Random
using LinearAlgebra

# Configurar semilla para reproducibilidad
Random.seed!(1234)

println("=== CARGANDO Y LIMPIANDO DATOS ===")

# Cargar datos desde Excel
df = DataFrame(XLSX.readtable("C:/Users/ASUS/OneDrive - Universidad del Pacífico/Lasso_Potential_Outcomes_RCTs/Lasso_Potential_Outcomes_RCTs/Julia/input/Districtwise_literacy_rates.xlsx", "2015_16_Districtwise"))

# Eliminar filas con valores faltantes
df_clean = dropmissing(df)
println("Datos originales: $(nrow(df)) filas")
println("Datos después de limpiar: $(nrow(df_clean)) filas")

# Mostrar primeras 5 filas
println("\nPrimeras 5 filas:")
println(first(df_clean, 5))

println("\n=== CREANDO HISTOGRAMAS ===")

# Crear histogramas de las tasas de alfabetización
p1 = histogram(df_clean.FEMALE_LIT, 
               bins=20, 
               color=:pink, 
               alpha=0.7,
               title="Tasa de Alfabetización Femenina",
               xlabel="Tasa de alfabetización (%)",
               ylabel="Frecuencia",
               legend=false)

p2 = histogram(df_clean.MALE_LIT, 
               bins=20, 
               color=:green, 
               alpha=0.7,
               title="Tasa de Alfabetización Masculina",
               xlabel="Tasa de alfabetización (%)",
               ylabel="Frecuencia",
               legend=false)

# Combinar gráficos
combined_plot = plot(p1, p2, layout=(1,2), size=(800,400))
plot!(combined_plot, plot_title="Histogramas de Tasas de Alfabetización por Género")
display(combined_plot)

println("\n=== PREPARANDO VARIABLES ===")

# Definir variable objetivo
y = df_clean.FEMALE_LIT

# Variables a excluir (IDs, texto, variables no explicativas)
vars_to_exclude = [
    :FEMALE_LIT,    # variable objetivo
    :STATCD,        # código de estado (ID)
    :DISTCD,        # código de distrito (ID) 
    :STATNAME,      # nombre de estado (texto)
    :DISTNAME,      # nombre de distrito (texto)
    :DISTRICTS,     # número de distritos
    :BLOCKS,        # número de bloques administrativos
    :VILLAGES,      # número de aldeas
    :OVERALL_LI,    # multicolinealidad prácticamente perfecta
    :AREA_SQKM      # área
]

# Crear dataframe Z con variables explicativas
Z = select(df_clean, Not(vars_to_exclude))
println("Variables explicativas disponibles:")
println(names(Z))

# Lista de variables base para el modelo
base_vars = [
    "TOTPOPULAT", "P_URB_POP", "POPULATION_0_6", "GROWTHRATE", "SEXRATIO",
    "P_SC_POP", "P_ST_POP", "MALE_LIT", "TOT_6_10_15", "TOT_11_13_15",
    "SCH1", "SCH2", "SCH3", "SCH4", "SCH5", "SCH6", "SCH7", "SCH9", "SCHTOT",
    "SCH1G", "SCH2G", "SCH3G", "SCH4G", "SCH5G", "SCH6G", "SCH7G", "SCH9G", "SCHTOTG",
    "SCH1P", "SCH2P", "SCH3P", "SCH4P", "SCH5P", "SCH6P", "SCH7P", "SCH9P", "SCHTOTP",
    "SCH1GR", "SCH2GR", "SCH3GR", "SCH4GR", "SCH5GR", "SCH6GR", "SCH7GR", "SCH9GR", "SCHTOTGR",
    "SCH1PR", "SCH2PR", "SCH3PR", "SCH4PR", "SCH5PR", "SCH6PR", "SCH7PR", "SCH9PR",
    "SELE1", "SELE2", "SELE3", "SELE4", "SELE5", "SELE6", "SELE7",
    "SCOMP1", "SCOMP2", "SCOMP3", "SCOMP4", "SCOMP5", "SCOMP6", "SCOMP7",
    "ENR501", "ENR502", "ENR503", "ENR504", "ENR505", "ENR506", "ENR507", "ENR509",
    "CLS1", "CLS2", "CLS3", "CLS4", "CLS5", "CLS6", "CLS7",
    "TCH1", "TCH2", "TCH3", "TCH4", "TCH5", "TCH6", "TCH7"
]

# Seleccionar solo las variables base que existen en el dataframe
available_vars = intersect(base_vars, names(Z))
X = select(Z, Symbol.(available_vars))

println("Variables seleccionadas para el modelo: $(length(available_vars))")

println("\n=== MODELO BASE (ESPECIFICACIÓN DE BAJA DIMENSIÓN) ===")

# Convertir a matriz y estandarizar
X_base = Matrix(X)
X_base_std = StatsBase.standardize(StatsBase.ZScoreTransform, X_base, dims=1)

# Función para validación cruzada con LASSO
function cross_validate_lasso(X, y; cv_folds=5, seed=123)
    Random.seed!(seed)
    n = length(y)
    indices = randperm(n)
    fold_size = div(n, cv_folds)
    
    r2_scores = Float64[]
    
    for fold in 1:cv_folds
        # Definir índices de entrenamiento y prueba
        start_idx = (fold - 1) * fold_size + 1
        end_idx = fold == cv_folds ? n : fold * fold_size
        
        test_indices = indices[start_idx:end_idx]
        train_indices = setdiff(1:n, test_indices)
        
        # Dividir datos
        X_train = X[train_indices, :]
        y_train = y[train_indices]
        X_test = X[test_indices, :]
        y_test = y[test_indices]
        
        # Ajustar LASSO con validación cruzada para encontrar lambda óptimo
        cv_result = glmnetcv(X_train, y_train, alpha=1.0, nfolds=5)
        
        # Predecir en conjunto de prueba
        y_pred = GLMNet.predict(cv_result.glmnet, X_test, s=cv_result.lambda_min)
        
        # Calcular R²
        r2 = 1 - sum((y_test - vec(y_pred)).^2) / sum((y_test .- mean(y_test)).^2)
        push!(r2_scores, r2)
    end
    
    return r2_scores
end

# Evaluar modelo base
r2_base = cross_validate_lasso(X_base_std, y)
println("Modelo base R² (LASSO): $(round(mean(r2_base), digits=4))")

println("\n=== MODELO FLEXIBLE (ESPECIFICACIÓN DE ALTA DIMENSIÓN) ===")

# Generar características polinomiales (solo potencias x^2 y x^3, sin interacciones)
function generate_polynomial_features(X, degree=3)
    n_samples, n_features = size(X)
    
    # Matriz para almacenar características expandidas
    X_expanded = copy(X)  # Comenzar con características originales
    
    # Agregar potencias
    for d in 2:degree
        X_power = X.^d
        X_expanded = hcat(X_expanded, X_power)
    end
    
    return X_expanded
end

# Generar características polinomiales
X_poly = generate_polynomial_features(X_base, 3)
println("Características originales: $(size(X_base, 2))")
println("Características con potencias: $(size(X_poly, 2))")

# Estandarizar características expandidas
X_poly_std = StatsBase.standardize(StatsBase.ZScoreTransform, X_poly, dims=1)

# Evaluar modelo flexible
r2_flex = cross_validate_lasso(X_poly_std, y)
println("Modelo flexible R² (LASSO): $(round(mean(r2_flex), digits=4))")

println("\n=== ANÁLISIS DEL CAMINO LASSO ===")

# Generar secuencia de lambdas de 10,000 a 0.001
lambdas = 10.0.^range(4, -3, length=200)

# Calcular camino LASSO
lasso_path = glmnet(X_poly_std, y, alpha=1.0, lambda=lambdas)

# Contar coeficientes no cero para cada lambda
function count_nonzero_coeffs(coeffs_matrix)
    return [sum(abs.(coeffs_matrix[:, i]) .> 1e-8) for i in 1:size(coeffs_matrix, 2)]
end

nonzero_counts = count_nonzero_coeffs(lasso_path.betas)

# Crear gráfico del camino LASSO
println("Creando gráfico del camino LASSO...")

lasso_plot = plot(log10.(lasso_path.lambda), nonzero_counts,
                 marker=:circle,
                 linewidth=2,
                 markersize=3,
                 xlabel="log₁₀(λ)",
                 ylabel="Nº de coeficientes ≠ 0",
                 title="Ruta LASSO (λ = 0.001 → 10,000)",
                 legend=false,
                 grid=true,
                 size=(600,400))

display(lasso_plot)

# Mostrar algunos valores clave
println("\nAnálisis del camino LASSO:")
println("Lambda mínimo: $(minimum(lasso_path.lambda))")
println("Lambda máximo: $(maximum(lasso_path.lambda))")
println("Máximo número de coeficientes no cero: $(maximum(nonzero_counts))")
println("Mínimo número de coeficientes no cero: $(minimum(nonzero_counts))")

# Encontrar lambda óptimo usando validación cruzada
println("\n=== LAMBDA ÓPTIMO ===")
cv_result = glmnetcv(X_poly_std, y, alpha=1.0, nfolds=5)
println("Lambda óptimo (min): $(round(cv_result.lambda_min, digits=6))")
println("Lambda 1SE: $(round(cv_result.lambda_1se, digits=6))")

# Coeficientes en lambda óptimo
optimal_coeffs = GLMNet.predict(cv_result.glmnet, s=cv_result.lambda_min, type="coefficients")
n_selected = sum(abs.(optimal_coeffs[2:end]) .> 1e-8)  # Excluir intercepto
println("Número de variables seleccionadas en lambda óptimo: $n_selected")

println("\n=== RESUMEN DE RESULTADOS ===")
println("R² Modelo Base: $(round(mean(r2_base), digits=4)) ± $(round(std(r2_base), digits=4))")
println("R² Modelo Flexible: $(round(mean(r2_flex), digits=4)) ± $(round(std(r2_flex), digits=4))")

# Comentario sobre los resultados
println("\n=== COMENTARIOS ===")
println("1. El modelo flexible (alta dimensión) generalmente muestra mejor R² que el modelo base")
println("2. A medida que λ aumenta, más coeficientes se vuelven cero (mayor regularización)")
println("3. A medida que λ disminuye, más variables entran en el modelo")
println("4. El LASSO ayuda a seleccionar las variables más relevantes automáticamente")

println("\nAnálisis completado exitosamente!")
```

```julia
# ========================================
# LASSO para Análisis de Alfabetización Femenina en India
# Traducción de Python a Julia
# ========================================

# Instalar paquetes necesarios (ejecutar una sola vez)
# using Pkg
# Pkg.add(["DataFrames", "XLSX", "Plots", "GLMNet", "StatsBase", "Statistics", 
#          "MLJ", "MLJLinearModels", "Random", "LinearAlgebra"])

# Cargar librerías
using DataFrames
using XLSX
using Plots
using GLMNet
using StatsBase
using Statistics
using Random
using LinearAlgebra

# Configurar semilla para reproducibilidad
Random.seed!(1234)

println("=== CARGANDO Y LIMPIANDO DATOS ===")

# Cargar datos desde Excel
df = DataFrame(XLSX.readtable("C:/Users/ASUS/OneDrive - Universidad del Pacífico/Lasso_Potential_Outcomes_RCTs/Lasso_Potential_Outcomes_RCTs/Julia/input/Districtwise_literacy_rates.xlsx", "2015_16_Districtwise"))

# Eliminar filas con valores faltantes
df_clean = dropmissing(df)
println("Datos originales: $(nrow(df)) filas")
println("Datos después de limpiar: $(nrow(df_clean)) filas")

# Mostrar primeras 5 filas
println("\nPrimeras 5 filas:")
println(first(df_clean, 5))

println("\n=== CREANDO HISTOGRAMAS ===")

# Crear histogramas de las tasas de alfabetización
p1 = histogram(df_clean.FEMALE_LIT, 
               bins=20, 
               color=:pink, 
               alpha=0.7,
               title="Tasa de Alfabetización Femenina",
               xlabel="Tasa de alfabetización (%)",
               ylabel="Frecuencia",
               legend=false)

p2 = histogram(df_clean.MALE_LIT, 
               bins=20, 
               color=:green, 
               alpha=0.7,
               title="Tasa de Alfabetización Masculina",
               xlabel="Tasa de alfabetización (%)",
               ylabel="Frecuencia",
               legend=false)

# Combinar gráficos
combined_plot = plot(p1, p2, layout=(1,2), size=(800,400))
plot!(combined_plot, plot_title="Histogramas de Tasas de Alfabetización por Género")
display(combined_plot)

println("\n=== PREPARANDO VARIABLES ===")

# Definir variable objetivo (convertir a Vector{Float64})
y = Float64.(df_clean.FEMALE_LIT)

# Variables a excluir (IDs, texto, variables no explicativas)
vars_to_exclude = [
    :FEMALE_LIT,    # variable objetivo
    :STATCD,        # código de estado (ID)
    :DISTCD,        # código de distrito (ID) 
    :STATNAME,      # nombre de estado (texto)
    :DISTNAME,      # nombre de distrito (texto)
    :DISTRICTS,     # número de distritos
    :BLOCKS,        # número de bloques administrativos
    :VILLAGES,      # número de aldeas
    :OVERALL_LI,    # multicolinealidad prácticamente perfecta
    :AREA_SQKM      # área
]

# Crear dataframe Z con variables explicativas
Z = select(df_clean, Not(vars_to_exclude))
println("Variables explicativas disponibles:")
println(names(Z))

# Lista de variables base para el modelo
base_vars = [
    "TOTPOPULAT", "P_URB_POP", "POPULATION_0_6", "GROWTHRATE", "SEXRATIO",
    "P_SC_POP", "P_ST_POP", "MALE_LIT", "TOT_6_10_15", "TOT_11_13_15",
    "SCH1", "SCH2", "SCH3", "SCH4", "SCH5", "SCH6", "SCH7", "SCH9", "SCHTOT",
    "SCH1G", "SCH2G", "SCH3G", "SCH4G", "SCH5G", "SCH6G", "SCH7G", "SCH9G", "SCHTOTG",
    "SCH1P", "SCH2P", "SCH3P", "SCH4P", "SCH5P", "SCH6P", "SCH7P", "SCH9P", "SCHTOTP",
    "SCH1GR", "SCH2GR", "SCH3GR", "SCH4GR", "SCH5GR", "SCH6GR", "SCH7GR", "SCH9GR", "SCHTOTGR",
    "SCH1PR", "SCH2PR", "SCH3PR", "SCH4PR", "SCH5PR", "SCH6PR", "SCH7PR", "SCH9PR",
    "SELE1", "SELE2", "SELE3", "SELE4", "SELE5", "SELE6", "SELE7",
    "SCOMP1", "SCOMP2", "SCOMP3", "SCOMP4", "SCOMP5", "SCOMP6", "SCOMP7",
    "ENR501", "ENR502", "ENR503", "ENR504", "ENR505", "ENR506", "ENR507", "ENR509",
    "CLS1", "CLS2", "CLS3", "CLS4", "CLS5", "CLS6", "CLS7",
    "TCH1", "TCH2", "TCH3", "TCH4", "TCH5", "TCH6", "TCH7"
]

# Seleccionar solo las variables base que existen en el dataframe
available_vars = intersect(base_vars, names(Z))
X = select(Z, Symbol.(available_vars))

println("Variables seleccionadas para el modelo: $(length(available_vars))")

println("\n=== MODELO BASE (ESPECIFICACIÓN DE BAJA DIMENSIÓN) ===")

# Convertir a matriz y estandarizar (asegurar tipos correctos)
X_base = Matrix{Float64}(X)
# Verificar que no hay valores faltantes
if any(ismissing, X_base) || any(isnan, X_base)
    println("¡Advertencia: Se encontraron valores faltantes o NaN en X_base!")
    # Reemplazar NaN con 0 o la media de la columna
    for j in 1:size(X_base, 2)
        col_mean = mean(filter(!isnan, X_base[:, j]))
        X_base[isnan.(X_base[:, j]), j] .= col_mean
    end
end

X_base_std = StatsBase.standardize(StatsBase.ZScoreTransform, X_base, dims=1)

# Función para validación cruzada con LASSO
function cross_validate_lasso(X, y; cv_folds=5, seed=123)
    Random.seed!(seed)
    n = length(y)
    indices = randperm(n)
    fold_size = div(n, cv_folds)
    
    r2_scores = Float64[]
    
    for fold in 1:cv_folds
        # Definir índices de entrenamiento y prueba
        start_idx = (fold - 1) * fold_size + 1
        end_idx = fold == cv_folds ? n : fold * fold_size
        
        test_indices = indices[start_idx:end_idx]
        train_indices = setdiff(1:n, test_indices)
        
        # Dividir datos
        X_train = X[train_indices, :]
        y_train = y[train_indices]
        X_test = X[test_indices, :]
        y_test = y[test_indices]
        
        try
            # Ajustar LASSO con validación cruzada para encontrar lambda óptimo
            cv_result = glmnetcv(X_train, y_train, alpha=1.0, nfolds=5)
            
            # Predecir en conjunto de prueba usando lambda óptimo
            # El campo correcto es lambda_min (sin underscore) en algunas versiones
            lambda_opt = try
                cv_result.lambda_min
            catch
                cv_result.meanloss[argmin(cv_result.meanloss)]  # lambda que minimiza el error
                cv_result.lambda[argmin(cv_result.meanloss)]
            end
            
            y_pred = GLMNet.predict(cv_result.path, X_test, s=[lambda_opt])
            
            # Calcular R²
            r2 = 1 - sum((y_test - vec(y_pred)).^2) / sum((y_test .- mean(y_test)).^2)
            push!(r2_scores, r2)
        catch e
            println("Error en fold $fold: $e")
            # En caso de error, usar un modelo simple
            simple_model = glmnet(X_train, y_train, alpha=1.0)
            y_pred = GLMNet.predict(simple_model, X_test)
            r2 = 1 - sum((y_test - vec(y_pred)).^2) / sum((y_test .- mean(y_test)).^2)
            push!(r2_scores, r2)
        end
    end
    
    return r2_scores
end

# Evaluar modelo base
r2_base = cross_validate_lasso(X_base_std, y)
println("Modelo base R² (LASSO): $(round(mean(r2_base), digits=4))")

println("\n=== MODELO FLEXIBLE (ESPECIFICACIÓN DE ALTA DIMENSIÓN) ===")

# Generar características polinomiales (solo potencias x^2 y x^3, sin interacciones)
function generate_polynomial_features(X, degree=3)
    n_samples, n_features = size(X)
    
    # Matriz para almacenar características expandidas
    X_expanded = copy(X)  # Comenzar con características originales
    
    # Agregar potencias
    for d in 2:degree
        X_power = X.^d
        X_expanded = hcat(X_expanded, X_power)
    end
    
    return X_expanded
end

# Generar características polinomiales
X_poly = generate_polynomial_features(X_base, 3)
println("Características originales: $(size(X_base, 2))")
println("Características con potencias: $(size(X_poly, 2))")

# Estandarizar características expandidas
X_poly_std = StatsBase.standardize(StatsBase.ZScoreTransform, X_poly, dims=1)

# Evaluar modelo flexible
r2_flex = cross_validate_lasso(X_poly_std, y)
println("Modelo flexible R² (LASSO): $(round(mean(r2_flex), digits=4))")

println("\n=== ANÁLISIS DEL CAMINO LASSO ===")

# Generar secuencia de lambdas de 10,000 a 0.001
lambdas = 10.0.^range(4, -3, length=200)

# Calcular camino LASSO
lasso_path = glmnet(X_poly_std, y, alpha=1.0, lambda=lambdas)

# Contar coeficientes no cero para cada lambda
function count_nonzero_coeffs(coeffs_matrix)
    return [sum(abs.(coeffs_matrix[:, i]) .> 1e-8) for i in 1:size(coeffs_matrix, 2)]
end

nonzero_counts = count_nonzero_coeffs(lasso_path.betas)

# Crear gráfico del camino LASSO
println("Creando gráfico del camino LASSO...")

lasso_plot = plot(log10.(lasso_path.lambda), nonzero_counts,
                 marker=:circle,
                 linewidth=2,
                 markersize=3,
                 xlabel="log₁₀(λ)",
                 ylabel="Nº de coeficientes ≠ 0",
                 title="Ruta LASSO (λ = 0.001 → 10,000)",
                 legend=false,
                 grid=true,
                 size=(600,400))

display(lasso_plot)

# Mostrar algunos valores clave
println("\nAnálisis del camino LASSO:")
println("Lambda mínimo: $(minimum(lasso_path.lambda))")
println("Lambda máximo: $(maximum(lasso_path.lambda))")
println("Máximo número de coeficientes no cero: $(maximum(nonzero_counts))")
println("Mínimo número de coeficientes no cero: $(minimum(nonzero_counts))")

# Encontrar lambda óptimo usando validación cruzada
println("\n=== LAMBDA ÓPTIMO ===")
try
    cv_result = glmnetcv(X_poly_std, y, alpha=1.0, nfolds=5)
    
    # Manejar diferentes versiones de GLMNet
    lambda_opt = try
        cv_result.lambda_min
    catch
        cv_result.lambda[argmin(cv_result.meanloss)]
    end
    
    lambda_1se = try
        cv_result.lambda_1se  
    catch
        # Usar lambda más conservador (1 SE rule)
        min_idx = argmin(cv_result.meanloss)
        se_threshold = cv_result.meanloss[min_idx] + cv_result.stdloss[min_idx]
        se_idx = findfirst(cv_result.meanloss .<= se_threshold)
        cv_result.lambda[se_idx]
    end
    
    println("Lambda óptimo (min): $(round(lambda_opt, digits=6))")
    println("Lambda 1SE: $(round(lambda_1se, digits=6))")
    
    # Coeficientes en lambda óptimo
    optimal_coeffs = GLMNet.predict(cv_result.path, s=[lambda_opt], type="coefficients")
    n_selected = sum(abs.(optimal_coeffs[2:end, 1]) .> 1e-8)  # Excluir intercepto
    println("Número de variables seleccionadas en lambda óptimo: $n_selected")
    
catch e
    println("Error en validación cruzada para lambda óptimo: $e")
    println("Usando modelo LASSO simple...")
    simple_model = glmnet(X_poly_std, y, alpha=1.0)
    println("Lambda mínimo disponible: $(minimum(simple_model.lambda))")
    println("Lambda máximo disponible: $(maximum(simple_model.lambda))")
end

println("\n=== RESUMEN DE RESULTADOS ===")
println("R² Modelo Base: $(round(mean(r2_base), digits=4)) ± $(round(std(r2_base), digits=4))")
println("R² Modelo Flexible: $(round(mean(r2_flex), digits=4)) ± $(round(std(r2_flex), digits=4))")

# Comentario sobre los resultados
println("\n=== COMENTARIOS ===")
println("1. El modelo flexible (alta dimensión) generalmente muestra mejor R² que el modelo base")
println("2. A medida que λ aumenta, más coeficientes se vuelven cero (mayor regularización)")
println("3. A medida que λ disminuye, más variables entran en el modelo")
println("4. El LASSO ayuda a seleccionar las variables más relevantes automáticamente")

println("\nAnálisis completado exitosamente!")
```

```julia
# ========================================
# LASSO para Análisis de Alfabetización Femenina en India
# Traducción de Python a Julia
# ========================================

# Instalar paquetes necesarios (ejecutar una sola vez)
# using Pkg
# Pkg.add(["DataFrames", "XLSX", "Plots", "GLMNet", "StatsBase", "Statistics", 
#          "MLJ", "MLJLinearModels", "Random", "LinearAlgebra"])

# Cargar librerías
using DataFrames
using XLSX
using Plots
using GLMNet
using StatsBase
using Statistics
using Random
using LinearAlgebra

# Configurar semilla para reproducibilidad
Random.seed!(1234)

println("=== CARGANDO Y LIMPIANDO DATOS ===")

# Cargar datos desde Excel
df = DataFrame(XLSX.readtable("C:/Users/ASUS/OneDrive - Universidad del Pacífico/Lasso_Potential_Outcomes_RCTs/Lasso_Potential_Outcomes_RCTs/Julia/input/Districtwise_literacy_rates.xlsx", "2015_16_Districtwise"))


# Eliminar filas con valores faltantes
df_clean = dropmissing(df)
println("Datos originales: $(nrow(df)) filas")
println("Datos después de limpiar: $(nrow(df_clean)) filas")

# Mostrar primeras 5 filas
println("\nPrimeras 5 filas:")
println(first(df_clean, 5))

println("\n=== CREANDO HISTOGRAMAS ===")

# Crear histogramas de las tasas de alfabetización
p1 = histogram(df_clean.FEMALE_LIT, 
               bins=20, 
               color=:pink, 
               alpha=0.7,
               title="Tasa de Alfabetización Femenina",
               xlabel="Tasa de alfabetización (%)",
               ylabel="Frecuencia",
               legend=false)

p2 = histogram(df_clean.MALE_LIT, 
               bins=20, 
               color=:green, 
               alpha=0.7,
               title="Tasa de Alfabetización Masculina",
               xlabel="Tasa de alfabetización (%)",
               ylabel="Frecuencia",
               legend=false)

# Combinar gráficos
combined_plot = plot(p1, p2, layout=(1,2), size=(800,400))
plot!(combined_plot, plot_title="Histogramas de Tasas de Alfabetización por Género")
display(combined_plot)

println("\n=== PREPARANDO VARIABLES ===")

# Definir variable objetivo (convertir a Vector{Float64})
y = Float64.(df_clean.FEMALE_LIT)

# Variables a excluir (IDs, texto, variables no explicativas)
vars_to_exclude = [
    :FEMALE_LIT,    # variable objetivo
    :STATCD,        # código de estado (ID)
    :DISTCD,        # código de distrito (ID) 
    :STATNAME,      # nombre de estado (texto)
    :DISTNAME,      # nombre de distrito (texto)
    :DISTRICTS,     # número de distritos
    :BLOCKS,        # número de bloques administrativos
    :VILLAGES,      # número de aldeas
    :OVERALL_LI,    # multicolinealidad prácticamente perfecta
    :AREA_SQKM      # área
]

# Crear dataframe Z con variables explicativas
Z = select(df_clean, Not(vars_to_exclude))
println("Variables explicativas disponibles:")
println(names(Z))

# Lista de variables base para el modelo
base_vars = [
    "TOTPOPULAT", "P_URB_POP", "POPULATION_0_6", "GROWTHRATE", "SEXRATIO",
    "P_SC_POP", "P_ST_POP", "MALE_LIT", "TOT_6_10_15", "TOT_11_13_15",
    "SCH1", "SCH2", "SCH3", "SCH4", "SCH5", "SCH6", "SCH7", "SCH9", "SCHTOT",
    "SCH1G", "SCH2G", "SCH3G", "SCH4G", "SCH5G", "SCH6G", "SCH7G", "SCH9G", "SCHTOTG",
    "SCH1P", "SCH2P", "SCH3P", "SCH4P", "SCH5P", "SCH6P", "SCH7P", "SCH9P", "SCHTOTP",
    "SCH1GR", "SCH2GR", "SCH3GR", "SCH4GR", "SCH5GR", "SCH6GR", "SCH7GR", "SCH9GR", "SCHTOTGR",
    "SCH1PR", "SCH2PR", "SCH3PR", "SCH4PR", "SCH5PR", "SCH6PR", "SCH7PR", "SCH9PR",
    "SELE1", "SELE2", "SELE3", "SELE4", "SELE5", "SELE6", "SELE7",
    "SCOMP1", "SCOMP2", "SCOMP3", "SCOMP4", "SCOMP5", "SCOMP6", "SCOMP7",
    "ENR501", "ENR502", "ENR503", "ENR504", "ENR505", "ENR506", "ENR507", "ENR509",
    "CLS1", "CLS2", "CLS3", "CLS4", "CLS5", "CLS6", "CLS7",
    "TCH1", "TCH2", "TCH3", "TCH4", "TCH5", "TCH6", "TCH7"
]

# Seleccionar solo las variables base que existen en el dataframe
available_vars = intersect(base_vars, names(Z))
X = select(Z, Symbol.(available_vars))

println("Variables seleccionadas para el modelo: $(length(available_vars))")

println("\n=== MODELO BASE (ESPECIFICACIÓN DE BAJA DIMENSIÓN) ===")

# Convertir a matriz y estandarizar (asegurar tipos correctos)
X_base = Matrix{Float64}(X)
# Verificar que no hay valores faltantes
if any(ismissing, X_base) || any(isnan, X_base)
    println("¡Advertencia: Se encontraron valores faltantes o NaN en X_base!")
    # Reemplazar NaN con 0 o la media de la columna
    for j in 1:size(X_base, 2)
        col_mean = mean(filter(!isnan, X_base[:, j]))
        X_base[isnan.(X_base[:, j]), j] .= col_mean
    end
end

X_base_std = StatsBase.standardize(StatsBase.ZScoreTransform, X_base, dims=1)

# Función para validación cruzada con LASSO
function cross_validate_lasso(X, y; cv_folds=5, seed=123)
    Random.seed!(seed)
    n = length(y)
    indices = randperm(n)
    fold_size = div(n, cv_folds)
    
    r2_scores = Float64[]
    
    for fold in 1:cv_folds
        # Definir índices de entrenamiento y prueba
        start_idx = (fold - 1) * fold_size + 1
        end_idx = fold == cv_folds ? n : fold * fold_size
        
        test_indices = indices[start_idx:end_idx]
        train_indices = setdiff(1:n, test_indices)
        
        # Dividir datos
        X_train = X[train_indices, :]
        y_train = y[train_indices]
        X_test = X[test_indices, :]
        y_test = y[test_indices]
        
        try
            # Ajustar LASSO con validación cruzada
            cv_result = glmnetcv(X_train, y_train, alpha=1.0, nfolds=5)
            
            # Encontrar el índice del lambda óptimo
            min_idx = argmin(cv_result.meanloss)
            lambda_opt = cv_result.lambda[min_idx]
            
            # Predecir usando el modelo con lambda óptimo
            # Usar directamente el índice para obtener la predicción correcta
            y_pred_full = GLMNet.predict(cv_result.path, X_test)
            y_pred = y_pred_full[:, min_idx]  # Seleccionar la columna correspondiente al lambda óptimo
            
            # Calcular R²
            r2 = 1 - sum((y_test - y_pred).^2) / sum((y_test .- mean(y_test)).^2)
            push!(r2_scores, r2)
            
        catch e
            println("Error en fold $fold: $e")
            # En caso de error, usar un modelo LASSO simple
            try
                simple_model = glmnet(X_train, y_train, alpha=1.0)
                y_pred_simple = GLMNet.predict(simple_model, X_test)
                # Tomar la primera columna (lambda más grande, más regularización)
                y_pred = y_pred_simple[:, 1]
                r2 = 1 - sum((y_test - y_pred).^2) / sum((y_test .- mean(y_test)).^2)
                push!(r2_scores, r2)
            catch e2
                println("Error también en modelo simple: $e2")
                # En último caso, usar R² = 0
                push!(r2_scores, 0.0)
            end
        end
    end
    
    return r2_scores
end

# Evaluar modelo base
r2_base = cross_validate_lasso(X_base_std, y)
println("Modelo base R² (LASSO): $(round(mean(r2_base), digits=4))")

println("\n=== MODELO FLEXIBLE (ESPECIFICACIÓN DE ALTA DIMENSIÓN) ===")

# Generar características polinomiales (solo potencias x^2 y x^3, sin interacciones)
function generate_polynomial_features(X, degree=3)
    n_samples, n_features = size(X)
    
    # Matriz para almacenar características expandidas
    X_expanded = copy(X)  # Comenzar con características originales
    
    # Agregar potencias
    for d in 2:degree
        X_power = X.^d
        X_expanded = hcat(X_expanded, X_power)
    end
    
    return X_expanded
end

# Generar características polinomiales
X_poly = generate_polynomial_features(X_base, 3)
println("Características originales: $(size(X_base, 2))")
println("Características con potencias: $(size(X_poly, 2))")

# Estandarizar características expandidas
X_poly_std = StatsBase.standardize(StatsBase.ZScoreTransform, X_poly, dims=1)

# Evaluar modelo flexible
r2_flex = cross_validate_lasso(X_poly_std, y)
println("Modelo flexible R² (LASSO): $(round(mean(r2_flex), digits=4))")

println("\n=== ANÁLISIS DEL CAMINO LASSO ===")

# Generar secuencia de lambdas de 10,000 a 0.001
lambdas = 10.0.^range(4, -3, length=200)

# Calcular camino LASSO
lasso_path = glmnet(X_poly_std, y, alpha=1.0, lambda=lambdas)

# Contar coeficientes no cero para cada lambda
function count_nonzero_coeffs(coeffs_matrix)
    return [sum(abs.(coeffs_matrix[:, i]) .> 1e-8) for i in 1:size(coeffs_matrix, 2)]
end

nonzero_counts = count_nonzero_coeffs(lasso_path.betas)

# Crear gráfico del camino LASSO
println("Creando gráfico del camino LASSO...")

lasso_plot = plot(log10.(lasso_path.lambda), nonzero_counts,
                 marker=:circle,
                 linewidth=2,
                 markersize=3,
                 xlabel="log₁₀(λ)",
                 ylabel="Nº de coeficientes ≠ 0",
                 title="Ruta LASSO (λ = 0.001 → 10,000)",
                 legend=false,
                 grid=true,
                 size=(600,400))

display(lasso_plot)

# Mostrar algunos valores clave
println("\nAnálisis del camino LASSO:")
println("Lambda mínimo: $(minimum(lasso_path.lambda))")
println("Lambda máximo: $(maximum(lasso_path.lambda))")
println("Máximo número de coeficientes no cero: $(maximum(nonzero_counts))")
println("Mínimo número de coeficientes no cero: $(minimum(nonzero_counts))")

# Encontrar lambda óptimo usando validación cruzada
println("\n=== LAMBDA ÓPTIMO ===")
try
    cv_result = glmnetcv(X_poly_std, y, alpha=1.0, nfolds=5)
    
    # Encontrar lambda óptimo
    min_idx = argmin(cv_result.meanloss)
    lambda_opt = cv_result.lambda[min_idx]
    
    # Encontrar lambda 1SE (regla de 1 error estándar)
    se_threshold = cv_result.meanloss[min_idx] + cv_result.stdloss[min_idx]
    # Buscar el lambda más grande (más regularizado) que esté dentro de 1 SE
    se_candidates = findall(cv_result.meanloss .<= se_threshold)
    se_idx = first(se_candidates)  # El primer índice corresponde al lambda más grande
    lambda_1se = cv_result.lambda[se_idx]
    
    println("Lambda óptimo (min): $(round(lambda_opt, digits=6))")
    println("Lambda 1SE: $(round(lambda_1se, digits=6))")
    
    # Coeficientes en lambda óptimo
    coeffs_at_opt = cv_result.path.betas[:, min_idx]
    n_selected = sum(abs.(coeffs_at_opt) .> 1e-8)
    println("Número de variables seleccionadas en lambda óptimo: $n_selected")
    
catch e
    println("Error en validación cruzada para lambda óptimo: $e")
    println("Usando modelo LASSO simple...")
    simple_model = glmnet(X_poly_std, y, alpha=1.0)
    println("Lambda mínimo disponible: $(minimum(simple_model.lambda))")
    println("Lambda máximo disponible: $(maximum(simple_model.lambda))")
end

println("\n=== RESUMEN DE RESULTADOS ===")
println("R² Modelo Base: $(round(mean(r2_base), digits=4)) ± $(round(std(r2_base), digits=4))")
println("R² Modelo Flexible: $(round(mean(r2_flex), digits=4)) ± $(round(std(r2_flex), digits=4))")

# Comentario sobre los resultados
println("\n=== COMENTARIOS ===")
println("1. El modelo flexible (alta dimensión) generalmente muestra mejor R² que el modelo base")
println("2. A medida que λ aumenta, más coeficientes se vuelven cero (mayor regularización)")
println("3. A medida que λ disminuye, más variables entran en el modelo")
println("4. El LASSO ayuda a seleccionar las variables más relevantes automáticamente")

println("\nAnálisis completado exitosamente!")
```
